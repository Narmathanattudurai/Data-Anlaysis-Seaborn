```python
import seaborn as sns
import matplotlip as plt
import pandas as pd
import numpy as np
```

```python
data=sns.load_dataset('tips')
```

```python
data.head(5)
```

```python
#scatter plot by relplot function
sns.relplot(x='total_bill', y='tip',data=data)
```

```python
#scatter plot with hue parameter
sns.relplot(x='total_bill', y='tip',hue='smoker',data=data)
```

```python
#scatter plot with style parameter
sns.relplot(x='total_bill', y='tip',hue='smoker', style='sex',data=data)
```

```python
#scatter plot with size parameter
sns.relplot(x='total_bill', y='tip',hue='smoker',style='sex',size='size',data=data)
```

```python
#strip plot
sns.catplot(x='day',y='tip',jitter=1,data=data)
```

```python
#strip plot with hue parameter
sns.catplot(x='day',y='tip',jitter=1,hue='sex',data=data)
```

```python
#Univariate analysis in box plot
sns.boxplot(x='tip',data=data)
```

```python
#multivariate analysis in box plot
sns.boxplot(x='day',y='total_bill',data=data)
```

```python
#Box plot with hue parameter
sns.boxplot(x='day',y='tip',hue='smoker',data=data)
```

```python
#violin plot
sns.violinplot(x='day',y='tip',data=data)
```

```python
#violin plot with hue parameter
sns.violinplot(x='day',y='tip',hue='sex',split=True,data=data)
```

```python
#Bar plots
sns.catplot(x='sex',y='tip',kind='bar',data=data)
```

```python
#bar plot with median values
sns.catplot(x='smoker',y='tip',kind='bar',estimator=np.median,data=data)
```

```python
#bar plot with standard deviation values
sns.catplot(x='smoker',y='tip',kind='bar',hue='sex',estimator=np.std,data=data)
```

```python
#countplots
sns.catplot(x='sex',kind='count',data=data)
```

```python
#countplots with hue parameter
sns.catplot(x='sex',kind='count',hue='smoker',data=data)
```

```python
#heatmap
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x=flights.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
#plotting heatmap
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5)
```

```python
sns.heatmap(x,cbar=False,linewidths=0.5,annot=True,fmt='d',cmap='winter')
```

```python
#clustermap
sns.clustermap(x)
```

```python
sns.clustermap(x,metric='correlation')
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=data)
```

```python
#jointplot with hexbin plot
sns.jointplot(x='total_bill',y='tip',kind='hex',data=data)
```

```python
#jointplot with kdeplot
sns.jointplot(x='total_bill',y='tip',kind='kde',data=data)
```

```python
#jointplot with regplot
sns.jointplot(x='total_bill',y='tip',kind='reg',data=data)
```

```python
sns.pairplot(iris)
```
